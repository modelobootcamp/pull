```python
# Pandas
import pandas as pd

# SQL Alchemy
from sqlalchemy import create_engine

```

```python
# Create a connection to the Census database

# Path to sqlite
database_path = "../Resources/Census_Data.sqlite"
# Create an engine that can talk to the database
engine = create_engine(f"sqlite:///{database_path}")
conn = engine.connect()
```

```python
# Query All Records in the the City Table


df = pd.read_sql("select * from Census_Data ; " , conn)

```

```python
df
```

```python
database_path = "../Resources/zip_census.sqlite"
zip_engine = create_engine(f"sqlite:///{database_path}")
zip_conn = zip_engine.connect()

zip_df = pd.read_sql("select * from Zip_Census; " , zip_conn)
```

```python
zip_df
```

```python
merge_df = pd.merge(df, zip_df, how = "inner", on="city")
```
